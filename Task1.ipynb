```python
#Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the newly uploaded dataset to check its contents
file_path_task1 = 'Task1.csv'
task1_data = pd.read_csv(file_path_task1)

# Show the first few rows of the dataset to understand its structure
task1_data.head(), task1_data.columns

```

```python

indicators = {
    'Country': ['China', 'India', 'United States', 'Indonesia', 'Brazil'],
    'GDP (current US$)': [17734063, 3176706, 23156540, 1195146, 2049414],
    'Life Expectancy (years)': [76.9, 70.8, 79.1, 71.7, 75.4],
    'Population (millions)': [1412600000, 1380000000, 331000000, 273000000, 212000000]
}

df = pd.DataFrame(indicators)

# Bar Graph
df.plot(x='Country', kind='bar', figsize=(8, 4), title='Comparison of Indicators')
plt.xticks(rotation=45)
plt.show()
```

```python
#Distribution of countries by GDP
plt.figure(figsize=(6, 4))
plt.hist(df['GDP (current US$)'], bins=6, color='green', edgecolor='black')
plt.title('Distribution of Countries by GDP')
plt.xlabel('GDP (current US$ in billions)')
plt.ylabel('Number of Countries')
plt.show()
```

```python
#population distribution
plt.figure(figsize=(5, 5))
plt.pie(df['Population (millions)'], labels=df['Country'], autopct='%1.1f%%', colors=['red', 'blue', 'green', 'orange', 'purple'])
plt.title('Population Distribution')
plt.show()
```

```python
# Example time-series data-population trend in India
years = ['2001', '2005', '2010', '2015', '2020']
china_pop = [1078970907, 1078970907,1322866505,1322866505, 1396387127]

plt.plot(years, china_pop, marker='o')
plt.title('Population Trend in India')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.show()
```

```python
#GDP vs Life expectancy
plt.figure(figsize=(6, 4))
plt.scatter(df['GDP (current US$)'], df['Life Expectancy (years)'], color='blue')

for i, txt in enumerate(df['Country']):
    plt.annotate(txt, (df['GDP (current US$)'][i], df['Life Expectancy (years)'][i]))

plt.title('GDP vs Life Expectancy')
plt.xlabel('GDP (current US$)')
plt.ylabel('Life Expectancy (years)')
plt.show()
```
